### Moda - Mediana - Media

image.png

**Media:** el promedio de un conjunto de n√∫meros. Para calcularla: sumamos todos los n√∫meros y  dividimos por la cantidad total de n√∫meros.
Ejemplo: 2, 4, 6 y 8, la media ser√≠a (2 + 4 + 6 + 8) / 4 = 5.

(La media da una idea de la tendencia central de los datos)


**Moda:** es el n√∫mero que aparece con mayor frecuencia en un conjunto de datos.
Se podria encontrar contando cu√°ntas veces aparece cada n√∫mero y seleccionando el que aparece con m√°s frecuencia.
Ejemplo: 2, 2, 4, 6 y 8, la moda ser√≠a 2

(La moda ayuda a identificar los valores m√°s comunes en los datos.)


**Mediana:** es el valor que ocupa la posici√≥n central en un conjunto de datos ordenados. Para calcularla, primero ordenamos los n√∫meros de menor a mayor y luego seleccionamos el n√∫mero en la posici√≥n central.
Si hay un n√∫mero impar de datos, la mediana es el n√∫mero en el medio.
Si hay un n√∫mero par de datos, la mediana es el promedio de los dos n√∫meros en el medio. Ejemplo:

2, 4, 6, 7 y 8, la mediana ser√≠a 6

2, 4, 6 y 8, la mediana ser√≠a 5 ((4 + 6) / 2 = 5)

(La mediana ayuda a entender c√≥mo se distribuyen los datos en relaci√≥n con su valor central.)

## MATPLOTLIB:

matplotlib.pyplot (plt) es una biblioteca en Python muy usada para la visualizaci√≥n de datos.  Genera gr√°ficos a partir de datos contenidos en listas o arrays

Permite crear una variedad de gr√°ficos, desde simples diagramas de dispersi√≥n hasta complejas visualizaciones tridimensionales


‚ûú Proporciona una API (Interfaz de Programaci√≥n de Aplicaciones): pylab, similar a la de MATLAB.

Matplotlib.pyplot es un m√≥dulo de la biblioteca Matplotlib que proporciona una interfaz gr√°fica de alto nivel para crear visualizaciones en Python. **Combina** las utilidades gr√°ficas del m√≥dulo **pyplot** con las capacidades num√©ricas de **NumPy**, lo que resulta en un entorno de trabajo que se asemeja a MATLAB en t√©rminos de manipulaci√≥n de matrices y vectores, y trazado de gr√°ficas.

Matplotlib.pyplot facilita la **creaci√≥n de gr√°ficos** mediante una interfaz simple y coherente. Permite a los usuarios crear figuras, subgr√°ficos y trazar l√≠neas, puntos, barras y m√°s, todo ello con una *sintaxis similar* a la de MATLAB. Adem√°s, proporciona una amplia gama de funciones para **personalizar** aspectos como colores, estilos, etiquetas y t√≠tulos de los gr√°ficos.

Al aprovechar las capacidades de NumPy, Matplotlib.pyplot puede manejar eficientemente datos num√©ricos en forma de matrices y vectores. Esto permite a los usuarios realizar operaciones matem√°ticas y estad√≠sticas en sus datos antes de trazarlos, lo que resulta en visualizaciones m√°s informativas y significativas.

### Limitaciones Matplotlib
- Interactividad limitada: Matplotlib est√° m√°s orientado a la creaci√≥n de gr√°ficos est√°ticos que a los gr√°ficos interactivos. Aunque se han desarrollado herramientas adicionales (como mplcursors) para agregar cierta interactividad, la funcionalidad interactiva no est√° integrada directamente en Matplotlib.

- Estilos predeterminados simples: Aunque Matplotlib ofrece una variedad de estilos predeterminados para los gr√°ficos, estos estilos pueden parecer simples en comparaci√≥n con algunas de las bibliotecas de visualizaci√≥n m√°s nuevas y modernas. Los gr√°ficos generados por defecto pueden requerir ajustes manuales para lograr una apariencia m√°s atractiva.

- Curva de aprendizaje empinada: Para usuarios nuevos, Matplotlib puede tener una curva de aprendizaje empinada debido a su amplia gama de funcionalidades y a la necesidad de comprender su estructura y sintaxis. La creaci√≥n de visualizaciones personalizadas y avanzadas puede requerir un conocimiento profundo de la biblioteca.

- Gr√°ficos 3D limitados: Aunque Matplotlib admite gr√°ficos 3D, la funcionalidad para gr√°ficos 3D es m√°s limitada en comparaci√≥n con otras bibliotecas especializadas en visualizaci√≥n tridimensional, como Mayavi o Plotly.

- Documentaci√≥n a veces confusa: A pesar de tener una documentaci√≥n extensa, la estructura y el formato de la documentaci√≥n de Matplotlib pueden ser confusos para algunos usuarios, lo que puede dificultar la b√∫squeda de soluciones a problemas espec√≠ficos.

- Flexibilidad en la personalizaci√≥n: Aunque Matplotlib es altamente personalizable, algunas personalizaciones avanzadas pueden requerir un c√≥digo adicional y pueden no ser tan intuitivas de implementar como en otras bibliotecas de visualizaci√≥n m√°s modernas. Esto puede dificultar la creaci√≥n de visualizaciones altamente personalizadas.

## SEABORN:
Seaborn es una biblioteca popular en Python que se utiliza principalmente para crear gr√°ficos estad√≠sticos de una manera simple y efectiva.

- Trazado estad√≠stico: Seaborn se especializa en la creaci√≥n de gr√°ficos estad√≠sticos, lo que significa que est√° dise√±ado para visualizar datos en t√©rminos de distribuciones, relaciones y comparaciones entre variables.

- Construido sobre Matplotlib: Seaborn se basa en la biblioteca de visualizaci√≥n de datos Matplotlib. Esto significa que puede aprovechar todas las funcionalidades de Matplotlib y, al mismo tiempo, proporciona una interfaz m√°s amigable y estilizada para la creaci√≥n de gr√°ficos.

- Estilos y paletas de colores predeterminados: Seaborn proporciona una variedad de estilos y paletas de colores predeterminados que pueden mejorar la apariencia de los gr√°ficos. Estos estilos y paletas est√°n dise√±ados para hacer que las visualizaciones sean m√°s atractivas y legibles.

- Integraci√≥n con DataFrames de Pandas: Seaborn est√° dise√±ado para funcionar de manera fluida con DataFrames de Pandas, una estructura de datos tabulares popular en Python. Esto hace que sea f√°cil y conveniente crear gr√°ficos a partir de datos almacenados en DataFrames.

### Limitaciones de Seaborn
- Complejidad limitada: Seaborn est√° dise√±ado principalmente para gr√°ficos estad√≠sticos y an√°lisis exploratorio de datos, por lo que puede carecer de la flexibilidad necesaria para crear visualizaciones altamente personalizadas o gr√°ficos complejos.

- Personalizaci√≥n limitada: Aunque Seaborn ofrece una variedad de estilos y paletas de colores predeterminados, la capacidad de personalizaci√≥n de estos estilos puede ser limitada en comparaci√≥n con bibliotecas m√°s flexibles como Matplotlib. Esto puede dificultar la creaci√≥n de gr√°ficos con un aspecto espec√≠fico.

- Documentaci√≥n limitada: Aunque Seaborn tiene una documentaci√≥n s√≥lida, puede no ser tan exhaustiva o detallada como la de otras bibliotecas de visualizaci√≥n de datos. Esto puede dificultar la resoluci√≥n de problemas o la comprensi√≥n de caracter√≠sticas menos utilizadas.

- Gr√°ficos espec√≠ficos: Seaborn se especializa en ciertos tipos de gr√°ficos estad√≠sticos, como diagramas de dispersi√≥n, diagramas de cajas y violines, y matrices de correlaci√≥n. Si necesitas crear otros tipos de gr√°ficos menos comunes, es posible que Seaborn no sea la mejor opci√≥n.

- Requerimientos de recursos: Algunas operaciones en Seaborn, especialmente aquellas que implican un gran n√∫mero de datos o gr√°ficos complejos, pueden ser intensivas en recursos y pueden llevar m√°s tiempo de procesamiento en comparaci√≥n con otras bibliotecas m√°s simples.

#### Plot:

La funci√≥n plot() se utiliza en Matplotlib para dibujar puntos (marcadores) en un diagrama. Por defecto, esta funci√≥n traza una l√≠nea que conecta los puntos, creando  un **gr√°fico de l√≠neas.**

El primer par√°metro de la funci√≥n plot() es una matriz que contiene los puntos del **eje x**, las coordenadas horizontales en el gr√°fico.

El segundo par√°metro de la funci√≥n plot() es una matriz que contiene los puntos del **eje y**, las coordenadas verticales en el gr√°fico. Ambos ejes pueden ser una lista de n√∫meros o un array NumPy que representa la ubicaci√≥n de los puntos de cada eje.

```python
# import matplotlib.pyplot as plt

# Datos para los ejes x - y
#x = [1, 2, 3, 4, 5]
#y = [2, 4, 6, 8, 10]
x = np.array ([1,8])
y = np.array ([3,10])

# Crear el gr√°fico
plt.plot(x, y)

# Etiquetas de los ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# T√≠tulo del gr√°fico - personalizar
plt.title('Gr√°fico de ejemplo')
#plt.plot(x, y, color='red', linewidth=5)
plt.plot(x, y, color='blue', linewidth=5, linestyle ="-.")

#plt.plot(x, y, color='green', linewidth=5, linestyle =':', label = 'Recta')
#plt.legend()

# Mostrar el gr√°fico
plt.show()
```

```python
# Datos
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Crear figura y ejes
fig, ax = plt.subplots()

# personalizar
ax.plot(x, y, color='pink', linewidth=5, linestyle='--', label='Recta')

# Agregar la leyenda con ubicaci√≥n en la parte superior y tama√±o de fuente 12
ax.legend(loc='upper left', fontsize=12)

# Etiquetas de los ejes
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')

# T√≠tulo del gr√°fico con tama√±o de fuente 16
ax.set_title('Gr√°fico con color, grosor y estilo personalizados', fontsize=16)

# Mostrar el gr√°fico
```

#### Distplot:

Distribuci√≥n de datos unidimensional: La distribuci√≥n de datos se refiere a c√≥mo est√°n distribuidos los valores en un conjunto de datos. Un conjunto de datos unidimensional significa que solo estamos analizando **una variable**.

 *distplot* combina un **histograma** y una **estimaci√≥n de la funci√≥n de densidad de kernel** (KDE) para visualizar la distribuci√≥n de datos. Un histograma muestra la frecuencia de los valores dentro de intervalos de valores (bins), mientras que la KDE es una forma suavizada de representar la distribuci√≥n de los datos.

```python
plt.figure()
sns.displot(blackfriday["Purchase"], color="#5ea88e", kde=True) #bins
# kde: estimaci√≥n de densidad del n√∫cleo, agrega una curva que representa la densidad de probabilidad de los datos
plt.xlabel('Monto de la compra') # eje
plt.ylabel('Frequency') # eje
sns.despine() # elimina las l√≠neas de los bordes superior y derecho - SNS
plt.show()
```

#### Boxplot:

o diagrama de caja y bigotes, es una forma simple pero efectiva de representar la distribuci√≥n de un conjunto de datos num√©ricos y visualizar las caracter√≠sticas clave de la distribuci√≥n

- Caja central: La caja en el centro del gr√°fico representa el rango intercuartil (IQR), que es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1) del conjunto de datos. El borde inferior de la caja (Q1) indica el valor por debajo del cual se encuentra el 25% de los datos, mientras que el borde superior (Q3) indica el valor por debajo del cual se encuentra el 75% de los datos.

- L√≠nea central: Dentro de la caja, hay una l√≠nea que indica la **mediana** del conjunto de datos, que es el valor que separa al 50% de los datos m√°s bajos del 50% de los datos m√°s altos.

- Bigotes (Whiskers): Los bigotes se extienden desde los bordes de la caja hasta los valores extremos del conjunto de datos. La longitud de los bigotes puede variar dependiendo de c√≥mo se defina el rango intercuartil y qu√© tan lejos se consideran los valores at√≠picos.

- Puntos at√≠picos: Los puntos individuales que se encuentran m√°s all√° de los bigotes se consideran valores at√≠picos y se pueden mostrar como puntos individuales en el gr√°fico.

*Interpretaci√≥n:* Un boxplot proporciona una manera r√°pida de identificar la dispersi√≥n y la simetr√≠a de los datos, as√≠ como cualquier valor at√≠pico. Tambi√©n permite comparar f√°cilmente varias distribuciones de datos en un solo gr√°fico.

#### Scaterplot:

 (o gr√°fico de dispersi√≥n) es una forma simple pero poderosa de visualizar la relaci√≥n entre dos conjuntos de datos num√©ricos.

- Puntos en un plano: gr√°ficos con un eje horizontal (x) y un eje vertical (y).

- Relaci√≥n entre variables: c√≥mo cambia una variable en funci√≥n de la otra.

Si los puntos se agrupan cerca de una l√≠nea diagonal, esto sugiere una relaci√≥n lineal positiva entre las variables. Si los puntos se alejan de esta l√≠nea, sugiere una relaci√≥n m√°s d√©bil o ninguna relaci√≥n en absoluto.

- Patrones y tendencias: Los scatterplots son √∫tiles para identificar patrones y tendencias en los datos (tendencia creciente o decreciente en los datos, o si hay agrupamientos o dispersi√≥n alrededor de ciertos valores).

- Personalizaci√≥n: color o el tama√±o de los puntos, agregando l√≠neas de regresi√≥n para mostrar la tendencia general de los datos, o resaltando grupos de puntos utilizando diferentes colores o formas.

Pueden trazar gr√°ficos bidimensionales que se pueden mejorar asignando hasta tres variables adicionales mientras se utiliza la
sem√°ntica de los par√°metros de tono, tama√±o y estilo.

#### Subplots

###### subplots
Para visualizar dos gr√°ficos (o mas) en una misma figura y compararlos, podemos utilizar la funci√≥n subplots de Matplotlib. Esta funci√≥n crea un objeto fig que representa toda la figura (es decir, el lienzo completo donde se realizan los gr√°ficos) y varios ejes en el objeto axes, que corresponden a los distintos subgr√°ficos que vamos a incluir dentro de la figura.

Por ejemplo, si queremos comparar dos gr√°ficos lado a lado, podemos dividir la figura en dos subgr√°ficos utilizando subplots y luego trazar nuestros datos en cada uno de los ejes de forma independiente.

```python
# Inicio subplots - FILAS - COLUMNAS
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(10, 4))

# primer grafico
ax[0].hist(blackfriday[blackfriday['City_Category'] == 'A']["Purchase"], color="#a8685e", alpha=0.5, bins=15)
# segundo grafico
ax[1].hist(blackfriday[blackfriday['City_Category'] == 'B']["Purchase"], color="#5ea88e", alpha=0.4, bins=20)
# tercer grafico
ax[2].hist(blackfriday[blackfriday['City_Category'] == 'C']["Purchase"], color="#0f0752", alpha=0.4,  bins=25)
# modifico primer grafico
ax[0].spines["top"].set_visible(False)
ax[0].spines["right"].set_visible(False)
# modifico segundo grafico
ax[1].spines["top"].set_visible(False)
ax[1].spines["right"].set_visible(False)
# modifico tercer grafico
ax[2].spines["top"].set_visible(False)
ax[2].spines["right"].set_visible(False)
plt.show()
```

```python
corr = blackfriday[['Age', 'Stay_In_Current_City_Years', 'Purchase']].corr(method = 'pearson')
```

Este c√≥digo calcula la **matriz de correlaci√≥n** entre las variables 'Age', 'Stay_In_Current_City_Years' y 'Purchase' del DataFrame blackfriday.

La matriz de correlaci√≥n es una tabla que muestra c√≥mo cada par de variables est√° relacionado entre s√≠.

En este caso, el m√©todo utilizado para calcular la correlaci√≥n es el coeficiente de **correlaci√≥n de Pearson**, que mide la fuerza y direcci√≥n de la relaci√≥n lineal entre dos variables continuas

**.corr(method='pearson')**: Calcula la correlaci√≥n entre estas variables usando el m√©todo de correlaci√≥n de Pearson. Ese m√©todo devuelve una matriz cuadrada donde los elementos de la diagonal principal son todos 1 (ya que la correlaci√≥n de una variable consigo misma es siempre perfecta), y los elementos fuera de la diagonal principal representan la correlaci√≥n entre pares de variables. Un valor cercano a 1 indica una correlaci√≥n positiva fuerte, un valor cercano a -1 indica una correlaci√≥n negativa fuerte, y un valor cercano a 0 indica una correlaci√≥n d√©bil o nula.


##### ¬øcomo vemos esa correlaci√≥n?


 heatmap es una visualizaci√≥n que utiliza colores para representar valores num√©ricos en una matriz. Funciona mostrando diferentes tonos de color en una cuadr√≠cula, donde cada celda de la cuadr√≠cula representa un par de valores en la matriz.

 annot=True agrega los valores num√©ricos en cada celda para facilitar la interpretaci√≥n.

```python
sns.heatmap(corr, cmap='YlGnBu', annot=True)
```

**Seaborn:**

- Seaborn es ideal para la visualizaci√≥n de datos estad√≠sticos y complejos.
- Funciona bien con DataFrames de Pandas y se integra perfectamente con las funcionalidades de Pandas.
- Es especialmente √∫til para crear gr√°ficos estad√≠sticos, como diagramas de dispersi√≥n con regresi√≥n, diagramas de caja, gr√°ficos de viol√≠n y gr√°ficos de distribuci√≥n conjunta.
- Seaborn proporciona una interfaz de alto nivel para crear gr√°ficos con estilos atractivos y colores predefinidos.

**Matplotlib:**

- Matplotlib es una biblioteca de bajo nivel que proporciona un control completo sobre los detalles de la visualizaci√≥n.
- Es √∫til cuando se necesita una personalizaci√≥n completa o cuando se quieren crear gr√°ficos que no est√°n disponibles en Seaborn.
- Ofrece una amplia gama de funciones para crear gr√°ficos 2D y 3D, incluyendo gr√°ficos de dispersi√≥n simples, gr√°ficos de l√≠neas, gr√°ficos de barras, gr√°ficos de torta, etc.
- Matplotlib es especialmente √∫til cuando se necesita un control preciso sobre aspectos como el tama√±o de la figura, la disposici√≥n de los ejes y las etiquetas, y la personalizaci√≥n de los estilos de l√≠nea y color.

**Combinaci√≥n:**
- Seaborn y Matplotlib pueden combinarse f√°cilmente en un mismo script.
- Seaborn utiliza Matplotlib por debajo, por lo que es posible personalizar a√∫n m√°s los gr√°ficos creados por Seaborn utilizando las funciones de Matplotlib.

Por ejemplo, se pueden agregar t√≠tulos, etiquetas de ejes adicionales, anotaciones y personalizaci√≥n de colores a los gr√°ficos de Seaborn utilizando las funciones de Matplotlib.

- Gr√°fico de barras para variables categ√≥ricas:

**sns.countplot:** para visualizar la distribuci√≥n de las variables categ√≥ricas como "Gender", "City_Category", "Marital_Status", etc.

- Gr√°fico de distribuci√≥n para variables num√©ricas:

**sns.histplot o sns.distplot:**  distribuci√≥n de variables num√©ricas como "Age", "Occupation", "Stay_In_Current_City_Years", etc.

- Gr√°fico de dispersi√≥n para relaciones entre variables num√©ricas:

**sns.scatterplot:** relaciones entre dos variables num√©ricas, como "Age" vs "Purchase", "Occupation" vs "Purchase", etc.

- Gr√°fico de caja para comparar distribuciones:

**sns.boxplot:** distribuci√≥n de una variable num√©rica entre diferentes categor√≠as, por ejemplo, "Purchase" en diferentes "City_Category".

- Mapas de calor para correlaci√≥n:

**sns.heatmap:** entre todas las variables num√©ricas en tu conjunto de datos

- ax.set_xlabel(titulo) : A√±ade un t√≠tulo con el contenido de la cadena titulo al eje x de ax. Se puede personalizar la alineaci√≥n y la fuente con los mismos par√°metros que para el t√≠tulo principal.
- ax.set_ylabel(titulo) : A√±ade un t√≠tulo con el contenido de la cadena titulo al eje y de ax. Se puede personalizar la alineaci√≥n y la fuente con los mismos par√°metros que para el t√≠tulo principal.
- ax.set_xlim([limite-inferior, limite-superior]) : Establece los l√≠mites que se muestran en el eje x de ax.
- ax.set_ylim([limite-inferior, limite-superior]) : Establece los l√≠mites que se muestran en el eje y de ax.
- ax.set_xticks(marcas) : Dibuja marcas en el eje x de ax en las posiciones indicadas en la lista marcas.
- ax.set_yticks(marcas) : Dibuja marcas en el eje y de ax en las posiciones indicadas en la lista marcas.
- ax.set_xscale(escala) : Establece la escala del eje x de ax, donde el par√°metro escala puede ser 'linear' (lineal) o 'log' (logar√≠tmica).
- ax.set_yscale(escala) : Establece la escala del eje y de ax, donde el par√°metro escala puede ser 'linear' (lineal) o 'log' (logar√≠tmica).

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3 ) * np.exp(-x ** 2 - y ** 2)

# Definir el n√∫mero de puntos a lo largo de cada eje
n = 10

# Crear arreglos de puntos a lo largo de los ejes x e y
x = np.linspace(-3, 3, int(3.5 * n))  # Crear 35 puntos a lo largo del eje x
y = np.linspace(-3, 3, int(3.0 * n))  # Crear 30 puntos a lo largo del eje y

# Crear una cuadr√≠cula de coordenadas X e Y utilizando meshgrid
X, Y = np.meshgrid(x, y)

# Calcular los valores Z (altura) para cada par de coordenadas (X, Y)
Z = f(X, Y)

# Crear una figura y ejes en las posiciones y tama√±os especificados
plt.axes([0.025, 0.025, 0.95, 0.95])

# Mostrar la matriz Z como una imagen
# interpolation='nearest' especifica que los p√≠xeles de la imagen se muestran sin interpolaci√≥n
# cmap='bone' especifica la paleta de colores que se utilizar√° para la imagen
# origin='lower' especifica que el origen de la imagen se encuentra en la esquina inferior izquierda
plt.imshow(Z, interpolation='nearest', cmap='bone', origin='lower')

# Agregar una barra de color asociada a los valores de Z en la imagen
# shrink=.92 especifica el factor de escala para la barra de color
plt.colorbar(shrink=.92)

# Eliminar los ticks del eje x e y
plt.xticks(())
plt.yticks(())

# Mostrar la visualizaci√≥n
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# Crear una nueva figura
fig = plt.figure()

# A√±adir un subplot 3D a la figura
ax = fig.add_subplot(111, projection='3d')

# Definir los valores de X e Y en el rango de -4 a 4 con un paso de 0.25
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)

# Crear una malla de coordenadas X e Y
X, Y = np.meshgrid(X, Y)

# Calcular el radio R a partir de las coordenadas X e Y
R = np.sqrt(X ** 2 + Y ** 2)

# Calcular los valores de Z a partir de la funci√≥n sin(R)
Z = np.sin(R)

# Crear una superficie 3D a partir de los datos de X, Y y Z
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.hot)

# A√±adir un gr√°fico de contorno en el plano Z = -2
ax.contourf(X, Y, Z, zdir='z', offset=-2, cmap=plt.cm.hot)

# Limitar el rango del eje z de -2 a 2
ax.set_zlim(-2, 2)

# Mostrar la figura
plt.show()
```
