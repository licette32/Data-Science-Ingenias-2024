### 1. Series

Una Serie es una estructura de datos de Pandas que representa una coleccin de elementos ordenados, como una lista. La diferencia principal es que cada elemento tiene una *etiqueta o ndice asociado*, lo que permite un acceso ms fcil a los datos y operaciones ms sofisticadas.

Una Serie de Pandas es similar a una lista o un array unidimensional en el sentido de que ambos contienen una secuencia ordenada de elementos.

**Diferencias:**
- indices personalizados (en el siguiente ejemplo vemos)
- Las Series tienen mtodos y funciones de Pandas integrados, lo que facilita la manipulacin, transformacin, y anlisis de datos. Esto incluye operaciones como filtrado, agrupacin y clculo de estadsticas descriptivas.
- Una Serie puede contener datos de diferentes tipos, como nmeros y cadenas de texto, mientras que un array de NumPy tiende a tener un tipo de dato uniforme (las listas pueden ser mixtas pero no tienen las mismas funciones)
- facilidad para tratar datos faltantes

````python
serie = pd.Series(data = [1,2,3, 4, 6.7,7,10])
````

````python
serie
````

- data: es un argumento que especifica los valores de la Serie
- index: especifica las etiquetas para cada uno de los valores en la Serie
- las etiquetas son ['primero', 'segundo', 'tercero', 'cuarto', 'quinto'], que es una lista de cadenas de texto
- las etiquetas se usan como ndices para los valores de la Serie. Para acceder a los valores podemos usar su posicin (0, 1, 2...), o su etiqueta ('primero', 'segundo', etc.).

### 2. Dataframes


Un DataFrame en Pandas es una estructura de datos bidimensional que se asemeja a una tabla de datos, similar a una hoja de clculo o una base de datos. Es una de las estructuras de datos ms importantes y verstiles en Pandas y se utiliza para almacenar y manipular datos de forma organizada.

Los DataFrames tienen dos dimensiones: filas y columnas.
Cada fila representa un registro o una observacin, mientras que cada columna representa una variable o atributo.

Las columnas de un DataFrame pueden contener diferentes tipos de datos (nmeros, cadenas de texto, fechas, etc.).
Esta flexibilidad permite trabajar con diferentes tipos de datos en una sola estructura.

Los DataFrames tienen muchos mtodos y funciones integrados para manipular, transformar y analizar datos, como filtrar, agrupar, fusionar, calcular estadsticas, etc.

8) A veces queremos seleccionar filas que cumplan con ciertas condiciones, donde el valor de una columna en esa fila sea igual, mayor o menor que un valor.

Para esto tenemos que usar una sintaxis especial. Vamos a construirla de a poco. Imaginemos que tenemos un DataFrame `df` con las columnas `col1`, `col2` y `col3`. Queremos seleccionar solo aquellas filas donde `col1` sea mayor a 10.

Para eso diremos que queremos

df['col1'] > 10 (La columna col1 debe ser mayor a 10).
Ahora si corremos este codigo, veremos que devuelve valores booleanos. O sea devolvera False para aquellos valores que sean menores o iguales a 10 y True para los que sean mayores a 10.

O sea que tenemos que agregar algo mas para poder seleccionar las columnas. Esta lista de valores booleanos se llama mascara booleana.

Que significa? Que si yo le paso estos valores a pandas, pandas interpretara que debe conservar aquellos valores donde tiene True y descartar donde tiene False.

Por eso, para filtrar filas en base a estas condiciones escribimos:

df[df['col1] > 10].

Esto significa primero fijate en que filas de `df`, la columna `col1` es mayor a 10. Luego, selecciona solo aquellas filas donde esta condicion sea `True`.
