###**DataFrame:**

Es la estructura de datos principal de Pandas y es similar a una tabla en una base de datos o una hoja de cálculo. Consiste en filas y columnas, donde cada columna puede tener un tipo de dato diferente (por ejemplo, enteros, flotantes, cadenas, etc.).

Los DataFrames permiten realizar operaciones de **manipulación** de datos como filtrado, clasificación, agrupación, y agregación de datos de forma eficiente.
También ofrecen una gran variedad de métodos para combinar y fusionar DataFrames, lo que facilita trabajar con múltiples conjuntos de datos.

###**Series:**

Es una estructura unidimensional que se parece a una columna de un DataFrame. Puede contener datos de cualquier tipo (números, cadenas, etc.) y tiene un índice que proporciona etiquetas para los datos.

Las Series son útiles para manipular datos como listas, pero con más funcionalidades. Por ejemplo, se pueden realizar operaciones aritméticas y estadísticas, filtrar valores, y más.

### 1. Series

Una Serie es una estructura de datos de Pandas que representa una colección de elementos ordenados, como una lista. La diferencia principal es que cada elemento tiene una *etiqueta o índice asociado*, lo que permite un acceso más fácil a los datos y operaciones más sofisticadas.

Una Serie de Pandas es similar a una lista o un array unidimensional en el sentido de que ambos contienen una secuencia ordenada de elementos.

**Diferencias:**
- indices personalizados (en el siguiente ejemplo vemos)
- Las Series tienen métodos y funciones de Pandas integrados, lo que facilita la manipulación, transformación, y análisis de datos. Esto incluye operaciones como filtrado, agrupación y cálculo de estadísticas descriptivas.
- Una Serie puede contener datos de diferentes tipos, como números y cadenas de texto, mientras que un array de NumPy tiende a tener un tipo de dato uniforme (las listas pueden ser mixtas pero no tienen las mismas funciones)
- facilidad para tratar datos faltantes

````python
serie = pd.Series(data = [1,2,3, 4, 6.7,7,10])
````

````python
serie
````

- data: es un argumento que especifica los valores de la Serie
- index: especifica las etiquetas para cada uno de los valores en la Serie
- las etiquetas son ['primero', 'segundo', 'tercero', 'cuarto', 'quinto'], que es una lista de cadenas de texto
- las etiquetas se usan como índices para los valores de la Serie. Para acceder a los valores podemos usar su posición (0, 1, 2...), o su etiqueta ('primero', 'segundo', etc.).

### 2. Dataframes


Un DataFrame en Pandas es una estructura de datos bidimensional que se asemeja a una tabla de datos, similar a una hoja de cálculo o una base de datos. Es una de las estructuras de datos más importantes y versátiles en Pandas y se utiliza para almacenar y manipular datos de forma organizada.

Los DataFrames tienen dos dimensiones: filas y columnas.
Cada fila representa un registro o una observación, mientras que cada columna representa una variable o atributo.

Las columnas de un DataFrame pueden contener diferentes tipos de datos (números, cadenas de texto, fechas, etc.).
Esta flexibilidad permite trabajar con diferentes tipos de datos en una sola estructura.

Los DataFrames tienen muchos métodos y funciones integrados para manipular, transformar y analizar datos, como filtrar, agrupar, fusionar, calcular estadísticas, etc.

````python
df = pd.DataFrame(data = [1,2,3, 4, 6.7],
          index=['a', 'b' ,'c', 'd', 'e'])
````

8) A veces queremos seleccionar filas que cumplan con ciertas condiciones, donde el valor de una columna en esa fila sea igual, mayor o menor que un valor.

Para esto tenemos que usar una sintaxis especial. Vamos a construirla de a poco. Imaginemos que tenemos un DataFrame `df` con las columnas `col1`, `col2` y `col3`. Queremos seleccionar solo aquellas filas donde `col1` sea mayor a 10.

Para eso diremos que queremos

df['col1'] > 10 (La columna col1 debe ser mayor a 10).
Ahora si corremos este codigo, veremos que devuelve valores booleanos. O sea devolvera False para aquellos valores que sean menores o iguales a 10 y True para los que sean mayores a 10.

O sea que tenemos que agregar algo mas para poder seleccionar las columnas. Esta lista de valores booleanos se llama mascara booleana.

¿Que significa? Que si yo le paso estos valores a pandas, pandas interpretara que debe conservar aquellos valores donde tiene True y descartar donde tiene False.

Por eso, para filtrar filas en base a estas condiciones escribimos:

df[df['col1] > 10].

Esto significa primero fijate en que filas de `df`, la columna `col1` es mayor a 10. Luego, selecciona solo aquellas filas donde esta condicion sea `True`.

````python
# Agrupacion de series
serie = pd.Series([1, 2, 3, 4, 5, 6, 7, 8, 9])
def es_par(num):
    return 'Par' if num % 2 == 0 else 'Impar'

# Agrupar la Serie usando la función anterior
grupo = serie.groupby(es_par)

# Mostrar los valores agrupados
for clave, valores in grupo:
    print(valores)
````
